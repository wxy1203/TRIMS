```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('TRIMS_April.xlsx')

total_rows = df.shape[0]
total_cols = df.shape[1]
print("Total number of rows:", total_rows)
print("Total number of cols:", total_cols)
```

```python
subtype_counts = df['incident_classification.subtype'].value_counts()
print("Number of sub-types:", len(subtype_counts))

subtype_others_count = subtype_counts.get("others", 0)
print("Number of 'others' incident sub-types:", subtype_others_count)
```

```python
rows_with_subtype_others = df[df['incident_classification.subtype'] == 'others']
selected_columns = rows_with_subtype_others[['incident_classification.category', 'incident_classification.type', 'incident_title']]
print(selected_columns)
rows_with_subtype_others['category_type'] = rows_with_subtype_others['incident_classification.category'] + ' - ' + rows_with_subtype_others['incident_classification.type'] 

# Count the occurrences of each category-type combination
category_type_counts = rows_with_subtype_others['category_type'].value_counts()

# Plot the bar chart
category_type_counts.plot(kind='bar')

# Add labels and title
plt.xlabel('Category - Type')
plt.ylabel('Occurrence Count')
plt.title('Occurrences of Category-Type Combinations')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right') 

# Show the plot
plt.show()
```

```python
df2 = pd.read_excel('Historical_Sitrep_22July_23Sep2023.xlsx', sheet_name='COMBINED_CLEAN')

total_rows = df2.shape[0]
total_cols = df2.shape[1]
print("Total number of rows:", total_rows)
print("Total number of cols:", total_cols)
```

```python


filtered_rows = df2[df2['Type'] == 'Others (Others)']

total_rows_filtered = filtered_rows.shape[0]


print("Total number of Others (Others) rows:", total_rows_filtered)

print("Others (Others) rows:", filtered_rows)
```

```python
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)
selected_columns = filtered_rows[['TRIMS_Category', 'TRIMS_Incident_Type']]
print(selected_columns)

grouped = selected_columns.groupby(['TRIMS_Category']).size().reset_index(name='occurrences')

# Plotting the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(range(len(grouped)), grouped['occurrences'], tick_label=grouped.apply(lambda x: f"{x['TRIMS_Category']}", axis=1))
plt.xticks(rotation=45, ha='right')
plt.xlabel('Category-Type - Incident-Type')
plt.ylabel('Occurrence')

plt.title('Occurrence of Category for SitRep Others (Others) Incident Type')

# Add value labels on top of each bar
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), round(bar.get_height(), 2), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
