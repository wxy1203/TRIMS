```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('TRIMS_April.xlsx')
print(df.iloc[0, :10])

total_rows = df.shape[0]
total_cols = df.shape[1]
print("Total number of rows:", total_rows)
print("Total number of cols:", total_cols)
```

```python
case_counts = df['case_list[0]'].value_counts()
print("Number of different cases:", len(case_counts))

filtered_case_counts = case_counts[case_counts >= 2]
print ("Numbwe of case occurring two times or more:", len(filtered_case_counts))
print("Number of items in each case (occurring two times or more):")
print(filtered_case_counts)

filtered_case_counts.plot(kind='bar')
plt.title('Occurrences of Cases (Two Times or More)')
plt.xlabel('Case')
plt.ylabel('Frequency')
plt.xticks(rotation=60)  
plt.show()
```
