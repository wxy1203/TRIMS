```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('Historical_Sitrep_22July_23Sep2023.xlsx', sheet_name=2)
print(df.iloc[0, :10])
```

```python
total_rows = df.shape[0]
total_cols = df.shape[1]
print("Total number of rows:", total_rows)
print("Total number of cols:", total_cols)
```

```python
dates = df['Date'].value_counts()
print("Number of different timestamps:", len(dates))


detail = df['What Happened'].value_counts()
print("Number of incident description:", len(detail))
# print("\nAll dates:")
# print(dates)
```

```python
type_counts = df['Type'].value_counts()
print("Number of different incident types:", len(type_counts))

print("\nNumber of items in each type:")
print(type_counts)

# Calculate the percentage of each type
percentages = 100 * type_counts / len(df)

# Filter out types with less than 2% occurrence
filtered_type_counts = type_counts[percentages >= 2]

plt.figure(figsize=(10, 6))
filtered_type_counts.plot(kind='pie', autopct=lambda p: '{:.1f}%'.format(p) if p >= 2 else '', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Distribution of Incident Types \n')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
location = df['Location'].value_counts()
print("Number of different locations:", len(location))

print("\nNumber of items in each location:")
print(location)

location = location[location >= 10]

plt.figure(figsize=(12, 6))
location.plot(kind='bar', color='skyblue')
plt.xlabel('Location')
plt.ylabel('Number of Occurrences')
plt.title('Distribution of Locations \n')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
cat_counts = df['TRIMS_Category'].value_counts()
print("Number of different incident categories:", len(cat_counts))

print("\nNumber of items in each category:")
print(cat_counts)


plt.figure(figsize=(10, 6))
cat_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Distribution of Incident Categories \n')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
t_counts = df['TRIMS_Incident_Type'].value_counts()
print("Number of different TRIMS incident types:", len(cat_counts))

print("\nNumber of items in each TRIMS type:")
print(t_counts)

# Calculate the percentage of each type
percentages = 100 * t_counts / len(df)

# Filter out types with less than 2% occurrence
filtered_type_counts = t_counts[percentages >= 1.5]

plt.figure(figsize=(10, 6))
filtered_type_counts.plot(kind='pie', autopct=lambda p: '{:.1f}%'.format(p) if p >= 1.5 else '', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Distribution of TRIMS Incident Types \n')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
